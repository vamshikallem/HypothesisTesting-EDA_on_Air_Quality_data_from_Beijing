# Dataset Description:

-This data set includes hourly air pollutants data from 12 nationally controlled air-quality monitoring sites. The air-quality data are from the Beijing Municipal Environmental Monitoring Center. The meteorological data in each air-quality site is matched with the nearest weather station from the China Meteorological Administration. The time period is from March 1st, 2013 to February 28th, 2017. Missing data are denoted as NA. The attributes are categorized in to three types which are indicated by different symbols in the proposal.

-Attribute Information:

* No: row number (Is Time series data)

* year: year of data in this row (Is Time series data)

* month: month of data in this row (Is Time series data)

* day: day of data in this row (Is Time series data)

* hour: hour of data in this row (Is Time series data)

* PM2.5: PM2.5 concentration (ug/m^3) (Is Gas concentrations in atmosphere)

* PM10: PM10 concentration (ug/m^3)   (Is Gas concentrations in atmosphere)

* SO2: SO2 concentration (ug/m^3)     (Is Gas concentrations in atmosphere)

* NO2: NO2 concentration (ug/m^3)     (Is Gas concentrations in atmosphere)

* CO: CO concentration (ug/m^3)       (Is Gas concentrations in atmosphere)

* O3: O3 concentration (ug/m^3)       (Is Gas concentrations in atmosphere)

+ TEMP: temperature (degree Celsius)   (Is Meteorological data at the same time)

+ PRES: pressure (hPa)                 (Is Meteorological data at the same time)

+ DEWP: dew point temperature (degree Celsius) (Is Meteorological data at the same time)

+ RAIN: precipitation (mm)                     (Is Meteorological data at the same time)

+ wd: wind direction                           (Is Meteorological data at the same time)

+ WSPM: wind speed (m/s)                       (Is Meteorological data at the same time)

+ station: name of the air-quality monitoring site (Is Meteorological data at the same time)
 



-The zip file consists of data collected from 12 different stations as 12 different csv files.

-Each file has 18 columns and 35000 rows and 2.7MB of data.

- Having different characteristic’s and missing values, there is good scope for Visualization and Data cleaning.

Hypothesis-4: 
HNull: Toxic gas concentrations increase over the years
HAlt: Toxic gas concentrations doesn't increase over the years

```python
DATA_season['year'].replace(to_replace=[2017],
    value=2013,
    inplace=True)
```

```python
Data_H4= DATA_season.groupby(['year'])["O3","CO",'NO2','SO2'].mean()
Data_H4.tail()
```

```python
fig, ax = plt.subplots(figsize=(150,50))
_=Data_H4.plot(kind = 'bar',ax=ax)
_=plt.xlabel('YEAR',fontsize=90)
_=plt.ylabel('Gas_Concentration(ug/m^3)',fontsize=90)
_=plt.title('Plot to compare Concentrations of all gases in different years',fontsize=100)
_=plt.xticks(rotation='vertical',fontsize=70)
_=plt.yticks(fontsize=70)
```

Hypothesis-4: Null Hypothesis is Rejected.

Gas Concentrations got reduced in year 2016 when compared to 2013.

```python
## using MANOVA test to know the dependency of columns on seasons
## NullH: columns of gas concentrations depend on seasons.
from statsmodels.multivariate.manova import MANOVA
DATA_season.columns = DATA_season.columns.str.replace(".", "_")
maov = MANOVA.from_formula('PM2_5 + PM10 + SO2 + NO2 + CO + O3  ~ season', data=DATA_season)
print(maov.mv_test())
```

here the pvalue is less than critical point value of 1.96 which means the null hypothesis is accepted.

```python
from scipy import stats
```

```python
## using test to know the difference in mean of gas conc. in diff seasons
winter_CO = DATA_season[DATA_season['season'] == 'winter']['CO']
summer_CO = DATA_season[DATA_season['season'] == 'summer']['CO']
spring_CO = DATA_season[DATA_season['season'] == 'spring']['CO']
fall_CO = DATA_season[DATA_season['season'] == 'fall']['CO']
stats.ttest_ind(winter_CO,summer_CO,equal_var = False) 
```

```python
stats.ttest_ind(winter_CO,fall_CO,equal_var = False)
```

in both the tests p vakue is lesser thancritical point value of 1.62 which accepts null hypotheses.
